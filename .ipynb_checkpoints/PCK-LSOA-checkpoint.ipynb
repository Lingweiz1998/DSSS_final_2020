## 1.0 | Introduction

This paper outlines the methodology and results of an exploratory cluster analysis of the labour market and public transport accessibility in London to understand the distributional characteristics and differences between the two dimensions in London.

Using census data and public transport indicator data from London Datastore, the paper proposes a combination of PCA downscaling and clustering to generate LSOA-level clustering results.

The results show that Outer and Inner London spatially show an unbalanced distribution of labour force and public transport. Outer London has a more concentrated labour force and more insufficient public transport accessibility, and Inner London has a more dispersed labour force and high public transport accessibility.

We also propose a conjecture about a behavioural pattern of workers living in Outer London and travelling to Inner London by public transport and suggest a possible experimental scenario for such a distribution.

## 2.0 | Literature review

### Literature

A labour market is a place where the market mechanism is used to regulate the supply and demand of labour, guide the proper flow of labour, and achieve a rational allocation of labour in all areas of the social economy (Investopedia, 2021). In recent years, research into different aspects of the labour market in cities has received much attention. Uneven distribution of labour can lead to uneven distribution of resources and over-concentration in cities. Several studies have explored the gender and ethnic imbalances in the labour market (Azarnova et al., 2020; Review of Social Economy, 2020; Shrestha et al., 2020). Many studies have analysed the uneven spatial distribution of the labour market (Li, Wei and Wu, 2019; Shrestha et al., 2020).

As one of the actions for sustainable development goals, the promotion of public transport has always been of interest. Many governments and academics have calculated the accessibility of public transport in cities using different algorithms, observed their distribution characteristics and made recommendations for improvement (Amaya, Cruzat and Munizaga, 2018; Saghapour, Moridpour and Thompson, 2016).

However, few studies have examined the distribution of the workforce in conjunction with the distribution of public transport. The Transit-Oriented-Development (TOD) perspective has become very popular (Mrsc.org, 2021). It suggests that using a transport hub as a centre for economic development to the periphery can bring people together around the hub, with the vast majority of trips being made by public transport (Vale, Viana and Pereira, 2018). This method minimises the consumption of resources and reduces the burden on roads (Vale, Viana and Pereira, 2018). This perspective compares and evaluates the development and public transport accessibility level and argues that good urban development should combine both economic development and public transport facilities.

By following this idea, this study analyses the distribution of the labour force and the distribution of public transport together, observing their distribution characteristics.

### Data and Research Question

We located the experimental area in London because of its rich and adequate database. We used London's Lower Super Output Area (LSOA) level database (LSOA Atlas), the minor regional classification for London boundaries. The LSOA Atlas database contains a large amount of census information and other indicator data, and we could easily find relevant data columns for the labour force and public transport.

Our research questions are therefore divided into two.

    RQ1 | What are the characteristics of London's labour force and public transport distribution?

    RQ2 | Is there a spatially uneven distribution of labour force and public transport development in London?

## 3.0 | Methodology

### Dimentionally Reduction and PCA

This study will use principal component analysis to reduce the dimensionality of the analysed objects and subsequently subject the reduced factors to k-means++ cluster analysis to observe these factors' distribution and clustering characteristics for economic and public transport.

In many research and application areas, it is often necessary to observe data containing multiple variables, collect large amounts of data, and then analyse them to find patterns. Large multivariate data sets can undoubtedly provide a wealth of information for research and applications, but they also increase the amount of data collection effort to some extent. More importantly, in many cases, many variables may be correlated with each other, increasing the complexity of the analysis of the problem. If each indicator is analysed separately, the analysis is often isolated and does not make full use of the data, so blindly reducing indicators can result in the loss of valuable information and erroneous conclusions.

Therefore, there is a need to find a reasonable method of reducing the number of indicators to be analysed while minimising the loss of information in the original indicators to achieve a comprehensive analysis of the data collected. Because there is a specific correlation between the variables, it is possible to consider turning the closely related variables into as few new variables as possible to be two-by-two uncorrelated. We can use a smaller number of composite indicators to represent the various types of information present in each variable separately. The principal component analysis is one of these dimensionality reduction algorithms.

PCA (Principal Component Analysis) is one of the most widely used algorithms for dimensionality reduction. The main idea of PCA is to map n-dimensional features to k-dimensions, which are new orthogonal features also known as principal components, reconstructed from the original n-dimensional features (Jolliffe and Cadima, 2016).

### K-means++ Clustering

Cluster analysis is the grouping of data objects based on the information found in the data that describes the objects and their relationships (Google Developers, 2020). The aim is that objects within groups are similar to each other (correlated), while objects in different groups are different (uncorrelated). The greater the similarity within groups and the more significant the gap between groups, the more significant the clustering is (Google Developers, 2020).

There are many different clustering algorithms, but K-Means is one of the most commonly used. It is simple, easy to understand and fast, but can only be applied to continuous data and must be manually specified before clustering (Xu and Tian, 2015).

The choice of initial values for k-means has a significant impact on the results, so improvements to initial values are essential (Xu and Tian, 2015). Of all the improvement algorithms, K-means++ is the best known. In simple terms, K-means++ is the selection of the point furthest from the chosen centroid.

### PCA Analysis

Next, We performed principal component analysis.

First, we need to choose a suitable number of components.

By looking at the change in the value of Eigenvalue to determine the suitable components.

```python
rand_st_int = 10 
pca = PCA(random_state=rand_st_int)

normed_pca = pca.fit_transform(normed)
print('Explained variance ratio of each component:')
print(pca.explained_variance_ratio_)
```

```python
dfpc = pd.DataFrame({'eigenvalue':pca.explained_variance_,
             'PC':list(range(1, pca.n_components_ + 1))})

plt.plot(dfpc['PC'], dfpc['eigenvalue'], marker='o', linestyle = '--')
plt.xlabel('Number of Components')
plt.ylabel('Eigenvalue')
plt.show()
```

As the diagram shows, the two points with the most significant transitions are 2 and 3, so a division into both 2 and 3 dimensions is appropriate.

As we envisaged in advance that the two objects of our study would be the `Employment` and `Public Transport`, we have chosen two components here.

```python
pca = PCA(n_components = 2)

normed_pca = pca.fit_transform(normed)
print('Explained variance ratio of each component:')
print(pca.explained_variance_ratio_)
```

```python
list_var = normed.columns.tolist()
print(list_var)
```

```python
df_PrincipleComp = pd.DataFrame(pca.components_, columns = list_var)
df_PrincipleComp
```

As shown in the table above, the larger `absolute value` in each column represents the factor to which that column belongs.

We can see that all the columns are very well grouped in the ideal aspect as we envisaged, except for `Eco_Active`, which has a similar interpretation rate in both factors and is therefore difficult to distinguish, but by summarising the other columns, we have been able to name the two components.

| Column      | Aspect |
|-------------|-----------|
| Eco_Active  | not significant| 
| Eco_inact   | Employment and Income   | 
| Employment  | Employment and Income   | 
| Unemployment| Employment and Income   |

| PTAL_Mean   | Public Transport inaccessability   |
| PTAL_good   | Public Transport inaccessability   | 
| PTAL_bad    | Public Transport inaccessability   |

```python
score_pca = pca.transform(normed)
score_pca
```

```python
print('Eigenvalues of each component:')
print(pca.explained_variance_)
```

### K-means++ Analysis

Next, we perform K-means++ clustering analysis.

First, as same with PCA, we decide on the number of clusters. We decide on the appropriate number of clusters by calculating the Sum of squared error, also known as SSE.

```python
# calculate SSE for a range of number of cluster
list_SSE = []
min_k = 1
max_k = 10
range_k = range(min_k, max_k+1)
for i in range_k:
    km = KMeans(
        n_clusters=i, init='k-means++', random_state=0)
    km.fit(score_pca)
    # inertia is a concept in physics. Roughly it means SSE of clustering.
    list_SSE.append(km.inertia_)

# plot
plt.plot(range_k, list_SSE, marker='o', linestyle = '--')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

Similarly, we found that the slope shift of the line graph was most significant in the range 2 to 4, so we decided to use 3 clusters.

```python
k_cluster = 3
random_seed = 1
kmeans_method = KMeans(n_clusters=k_cluster,random_state=random_seed)
kmeans_method.fit(score_pca)
```

### Result (Data Visualization)

When we projected the clustering results onto a London map, we found that these clusters also had some characteristics in their spatial distribution.

As shown, two clusters are concentrated in `Inner London`, while one cluster is concentrated in `Outer London`.

```python
ppd['Kmeans_Cluster'] = kmeans_method.labels_
ppd.plot(column='Kmeans_Cluster', categorical=True, legend=True, figsize=(12,8));
```

```python
ppd_kpca = pd.concat([ppd.reset_index(drop = True), pd.DataFrame(score_pca)],axis = 1)
ppd_kpca.columns.values[-2: ] = ['Component1','Component2']
```

```python
ppd_kpca['Segement'] = ppd_kpca['Kmeans_Cluster'].map({0:'First',1:'Second',2:'Third'})
```

```python
x_axis = ppd_kpca['Component1']
y_axis = ppd_kpca['Component2']
plt.figure(figsize = (10, 8))
sns.scatterplot(x_axis,y_axis,hue = ppd_kpca['Segement'], palette = ['g','r','c'])
plt.title('Clusters by PCA Components')
plt.show()
```

## 6.0 | Discussion and Conclusion

In conclusion, this study has identified several spatially distributed characteristics of the labour force distribution and public transport accessibility in London. 

    In Outer London, transport accessibility is poor, with employment rates and income levels performing better. 

    In Inner London, although there are two distinct clusters, there is generally better performance in transport accessibility, with more unsatisfactory performance in employment rates and income levels in one of the clusters. 

This illustrates the uneven distribution pattern of London's Labors and public transport accessibility across Inner London and Outer London.

We speculate that although the workforce is more concentrated in Outer London, this does not mean that there are more jobs available in these areas. The pattern may have something to do with the higher cost of housing in Inner London, where many workers choose to live in Outer London and then travel to Inner London to work. The ability to commute daily to the city centre for work is also evidence of Inner London's better public transport accessibility.

However, these are only speculations, and more analysis is needed to verify this conjecture. Therefore, the following analyses should include relevant data such as house prices and the number of jobs on offer.
